# Model Building

```python
from sklearn.model_selection import train_test_split
```

```python
s=data.drop("SalePrice",axis=1)
p=data["SalePrice"]
```

```python
s_train,s_test,p_train,p_test=train_test_split(s,p,test_size=0.2,random_state=42)
```

```python
print(s_train.shape)
print(s_test.shape)
print(p_train.shape)
print(p_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
lr =LinearRegression()
```

```python
lr.fit(s_train,p_train)
```

### Predictions

```python
predictions=lr.predict(s_test)
print("Actual val",p_test[0])
print("Model predict val",predictions[0])
```

```python
from sklearn.metrics import mean_squared_error
mse= mean_squared_error (p_test,predictions)
rmse=np.sqrt(mse)
print(rmse)
```
