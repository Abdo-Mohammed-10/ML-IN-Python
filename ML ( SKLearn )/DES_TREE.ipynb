# Decision Tree  ( classification & regression )

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
from sklearn import datasets,model_selection,linear_model,metrics,preprocessing,tree
from sklearn.model_selection import train_test_split
```

```python
x,y=datasets.load_iris(return_X_y=True)
x_normalize=preprocessing.StandardScaler()
x_norm=x_normalize.fit_transform(x)
x.shape,y.shape,x_norm.shape
```

```python
x_train,x_test,y_train,y_test=model_selection.train_test_split(x_norm,y,test_size=0.1,random_state=42,stratify=y)
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
DT=tree.DecisionTreeClassifier()
DT.fit(x_train,y_train)
```

```python
DT.predict(x_test)
```

```python
metrics.accuracy_score(y_test,y_pred=DT.predict(x_test))
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,10),dpi=80)
tree.plot_tree(DT)
```
